```python
# - Change time format of both files to Pandas datetime// 


import pandas as pd 
dfA= pd.read_csv("C:/Users/kidist/Downloads/Electricity_consumption_2015-2025 (1).csv")
display (dfA.head())
dfB= pd.read_csv("C:/Users/kidist/Downloads/Electricity_price_2015-2025.csv",  delimiter = ";")
display (dfB.head(10))
```

```python
dfA["datetime"]= pd.to_datetime(dfA["time"], format = "%Y-%m-%d %H:%M:%S" )
display(dfA.head(5))
dfB["datetime"] = pd.to_datetime(dfB["timestamp"], format = "%H:%M %m/%d/%Y")
display(dfB.head(4))
```

```python
# - Join the two data frames according to time //

df= pd.merge(dfA, dfB, on='datetime' , how = 'inner')
df
```

```python
# - Calculate the hourly bill paid (using information about the price and the consumption) //


df["Price"] = df["Price"].astype(str).str.replace(',', '.').astype(float)  #changing the price decimal separator
#display(df.head())
df["hourlybill"]= (df["Price"]/100 )* df["kWh"]   #hourlybill is in (euros)
display(df[["hourlybill", "datetime", "Price", "kWh"]])
```

```python
# - Calculated grouped values of daily, weekly or monthly consumption, bill, average price and average temperature //


df_avg_daily= df.groupby(pd.Grouper(key='datetime', freq= 'D')).agg({"kWh": "sum", "hourlybill": "sum", "Price":"mean", "Temperature":"mean"}).reset_index() 
df_avg_weekly= df.groupby(pd.Grouper(key='datetime', freq= 'W')).agg({"kWh": "sum", "hourlybill" : "sum", "Price" :"mean", "Temperature":"mean"}).reset_index() 
df_avg_monthly= df.groupby(pd.Grouper(key='datetime', freq= 'ME')).agg({"kWh": "sum", "hourlybill" : "sum", "Price" :"mean", "Temperature":"mean"}).reset_index() 

display(df_avg_daily)
display(df_avg_weekly)
display(df_avg_monthly)
```

```python
df_avg_daily.to_csv('daily_avg_data.csv', index=False)
df_avg_weekly.to_csv('weekly_avg_data.csv', index=False)
df_avg_monthly.to_csv('monthly_avg_data.csv', index=False)
```
